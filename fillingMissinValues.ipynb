```python
import pandas as pd
df = pd.read_csv("nyc_weather.csv", parse_dates=['EST'])
df.set_index('EST', inplace=True)
df.head()
```

```python
#Filling missing values with Zeroes
new_df = df.fillna(0)
```

```python
new_df.head()
```

```python
#Filling missing values with dictionary
new_df = df.fillna({
    'Temperature': 0,
    'WindSpeedMPH' : 0,
    'Events': 'No Event'
})
new_df.head()
```

```python
#Filling missing values with 'Forward Fill'
new_df = df.fillna(method='bfill')
new_df
```
