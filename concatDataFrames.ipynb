```python
import pandas as pd
india_weather = pd.DataFrame({
    'cities' : ['Mumbai', 'Delhi', 'Banglore'],
    'Temperature' : [32, 45, 30],
    'Humidity' : [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    'cities' : ['New York', 'Chicago', 'Orlando'],
    'Temperature' : [21, 14, 35],
    'Humidity' : [65, 65, 75]
})
us_weather
```

```python
#Using concat method to append two dataframes
df = pd.concat([india_weather, us_weather])
```

```python
df
```

```python
#Using ignore index to get the continuous index
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
#Using Keys parameter
df = pd.concat([india_weather, us_weather], keys = ["India", "US"])
df
```

```python
#Using loc instance to retrieve a subset dataframe 
df.loc["India"]
```

```python
df.loc["US"]
```

```python
# Using  to append a new dataframe as a column
temperature_df = pd.DataFrame({
    'cities' : ['Mumbai', 'Delhi', 'Banglore'],
    'Temperature' : [32, 45, 30],
})
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'cities' : ['Mumbai', 'Delhi', 'Banglore'],
    'windspeed' : [7, 12, 9]
})
windspeed_df
```

```python
new_df = pd.concat([temperature_df, windspeed_df], axis=1)
new_df
```

```python
# Appending a series to a new dataframe
#lets create a simple series
s = pd.Series(["Humid", "Dry", "Rain"], name = "Event")
s
```

```python
s_df = pd.concat([temperature_df, s], axis=1)
s_df
```
